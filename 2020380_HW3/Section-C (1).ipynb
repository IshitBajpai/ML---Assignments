```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
import seaborn as sns


epochs = 50
batch_size = 140

train_loss = []
val_loss = []

model = MLPClassifier(hidden_layer_sizes=(256,32),activation='logistic',max_iter=1)
for i in range(epochs):
    idx = 0
    number_of_batches = int(len(X_train)/batch_size)
    for j in range(number_of_batches):
        model.partial_fit(X_train[j*batch_size:(j+1)*(batch_size),:],y_train[j*batch_size:(j+1)*(batch_size)],classes=np.unique(y_train))
    train_loss.append(log_loss(y_train,model.predict_proba(X_train)))
    val_loss.append(log_loss(y_val,model.predict_proba(X_val)))
   


print("Training Loss",train_loss[-1])
print("Validation Loss",val_loss[-1])

sns.lineplot(x=np.arange(1,50+1,1),y=train_loss,label='Train')
sns.lineplot(x=np.arange(1,50+1,1),y=val_loss,label='Val')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
import seaborn as sns

best_activation = 'relu'
learning_rate = [0.1,0.01,0.001]

Train_loss = []
Val_loss = []

for k in range(0,len(learning_rate)):
  
    epochs = 50
    batch_size = 200

    train_loss = []
    val_loss = []

    model = MLPClassifier(hidden_layer_sizes=(256,32),activation='relu',max_iter=100,learning_rate_init=learning_rate[k])
    for i in range(epochs):
        idx = 0
        number_of_batches = int(len(X_train)/batch_size)
        for j in range(number_of_batches):
            model.partial_fit(X_train[j*batch_size:(j+1)*(batch_size),:],y_train[j*batch_size:(j+1)*(batch_size)],classes=np.unique(y_train))
        train_loss.append(log_loss(y_train,model.predict_proba(X_train)))
        val_loss.append(log_loss(y_val,model.predict_proba(X_val)))


    print("alpha = ",learning_rate[k])
    print("Training Loss",train_loss[-1])
    print("Validation Loss",val_loss[-1])
    print("--------------------------")
    
    Train_loss.append(train_loss)
    Val_loss.append(val_loss)    
```

```python
sns.lineplot(x=np.arange(1,50+1,1),y=Train_loss[0],label='0.1',color='red')
sns.lineplot(x=np.arange(1,50+1,1),y=Train_loss[1],label='0.01',color='green')
sns.lineplot(x=np.arange(1,50+1,1),y=Train_loss[2],label='0.001',color='blue')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()
```

```python
sns.lineplot(x=np.arange(1,50+1,1),y=Val_loss[0],label='0.1',color='red',linewidth = 2)
sns.lineplot(x=np.arange(1,50+1,1),y=Val_loss[1],label='0.01',color='green')
sns.lineplot(x=np.arange(1,50+1,1),y=Val_loss[2],label='0.001',color='blue')

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Validation Loss")
plt.legend()
```

# Q4

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
import seaborn as sns

model = MLPClassifier(hidden_layer_sizes=(256,32),max_iter=100,early_stopping=True)
parameters = {
    
    'solver': ['adam','sgd'],
    'learning_rate': ['adaptive','constant'],
    'learning_rate_init': [0.001, 0.01,0.1],
    'activation': ['logistic','tanh', 'relu','identity']
}                
```

```python
from sklearn.model_selection import GridSearchCV
best_model = GridSearchCV(model,parameters, n_jobs=-1, cv=3)
best_model.fit(X_train, y_train) 
```

```python
print('Best parameters \n', best_model.best_params_)
```

```python
best_model.score(X_train,y_train)
```

```python
best_model.score(X_val,y_val)
```

```python
print(log_loss(y_train,best_model.predict_proba(X_train)))
print(log_loss(y_val,best_model.predict_proba(X_val)))
```
